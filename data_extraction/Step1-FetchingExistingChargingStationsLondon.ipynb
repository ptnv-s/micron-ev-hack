```python
import pandas as pd
```

```python
# cumulative dataset from OpenChargeMap
df = pd.read_json('datasets_in/poi.json', lines=True)
```

```python
df
```

```python
# taking relevent information
df_new = df[['AddressInfo', 'Connections', 'OperatorInfo', 'SpatialPosition']]
```

```python
# convert point geometry to latitude longitude (decimal degrees)
latitude = []
longitude = []
for i in df_new['SpatialPosition']:
  coordinates = i.get('coordinates')
  longitude.append(coordinates[0])
  latitude.append(coordinates[1])
```

```python
df_new['latitude'] = pd.Series(latitude)
df_new['longitude'] = pd.Series(longitude)
```

```python
df_new
```

```python
# filtering the dataset to only include data about charging stations in London. (using bounding box coordinates of London)
df_new = df_new[df_new['latitude']>51.274519]
df_new = df_new[df_new['latitude']<51.682778]
df_new = df_new[df_new['longitude']>-0.533878]
df_new = df_new[df_new['longitude']<0.245327]
```

```python
df_new = df_new[['AddressInfo', 'OperatorInfo', 'latitude', 'longitude']]
```

```python
df_new.to_csv('datasets_out/london_cs_data.csv')
```

```python
# using folium to plot existing charging  stations
import folium
import tqdm


this_map = folium.Map(prefer_canvas=True)

def plotDotCS(latitude,longitude,color, this_map):
  
    folium.CircleMarker(location=[latitude, longitude],
                        radius=0.5,
                        weight=5,
                        color=color).add_to(this_map)



for latitude, longitude in zip(df_new['latitude'], df_new['longitude']):
  plotDotCS(latitude, longitude, 'blue', this_map)
   


this_map.fit_bounds(this_map.get_bounds())


this_map
```

```python
# fetching data using OpenStreetMap about regions having listed amenities in London 
import osmnx as ox
tags_not_cs = {'amenity': ['food_court', 'restaurant', 'college', 'university', 'parking', 'hospital']} #amenities
gdf_london_other = ox.geometries.geometries_from_bbox(north=51.682778, south=51.274519, east=0.245327, west=-0.533878, tags=tags_not_cs)
gdf_london_other.shape
```

```python
gdf_london_other
```

```python
# only retaining point features for simplicity
gdf_london_other = gdf_london_other[gdf_london_other.geom_type != 'Polygon']
gdf_london_other = gdf_london_other[gdf_london_other.geom_type != 'MultiPolygon']
```

```python
gdf_london_other = gdf_london_other[['amenity', 'name', 'geometry']]
```

```python
# encoding the amenities as 0, 1
one_hot = pd.get_dummies(gdf_london_other['amenity'])
gdf_london_other = gdf_london_other.drop('amenity', axis=1)
gdf_london_other = gdf_london_other.join(one_hot)
```

```python
gdf_london_other.to_csv('datasets_out/other_amenities.csv')
gdf_london_other
```
